```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mpl_toolkits.mplot3d as a_3D
```

```python
def subs(v1,v2):
    return v1 - v2
```

```python
def dot(v1,v2):
    return np.dot(v1,v2)
```

```python
def sum(v1,v2):
    return v1 + v2
```

```python
# Вход от потребителя за 3D вектори
v1_input = input("Въведи елементите на вектор v1 (напр. 1 2 3): ")
v2_input = input("Въведи елементите на вектор v2 (напр. 4 5 6): ")

# Разделяне и преобразуване на входа в числа
v1 = pd.array([float(x.strip()) for x in v1_input.replace(',',' ').split()])
v2 = pd.array([float(x.strip()) for x in v2_input.replace(',',' ').split()])
```

### Това е сумата на два вектора!

```python
print(sum(v1,v2).tolist())
```

### Това е скаларен продукт!

```python
print(dot(v1,v2))
```

### Това е разлика на два вектора!

```python
print(subs(v1,v2).tolist())
```

### Taка нормализираме вектор

```python
print(np.linalg.norm(v1))
```

### Така правим скаларно умножение на вектори

```python
scalar = 2
v1_scalar = scalar * v1
print(v1_scalar.tolist())
```

### Чертаем векторите и да покажем операциите



* самите вектори - 3D
* quiver - работи само за 2D - вектори
* за 3D използваме mpl_toolkits.mplot3d 

```python
#Настройка на графиката
fig = plt.figure()
ax= fig.add_subplot(111,projection = '3d')


ax.quiver(0,0,0, v1[0],v1[1],v1[2], color='r', label = 'v1')
ax.quiver(0,0,0, v2[0],v2[1],v2[2], color='b', label = 'v2')

ax.set_xlim(0, max(v1[0], v2[0]) + 1)
ax.set_ylim(0, max(v1[1], v2[1]) + 1)
ax.set_zlim(0,max(v1[2],v2[2]) + 1)

ax.set_xlabel('X ос')
ax.set_ylabel('Y ос')
ax.set_zlabel('Z ос')
ax.legend()
ax.set_title('3D Вектори v1 и v2')
plt.show()

```

* Сумата на двата вектора, но в 2D


```python
plt.quiver(0, 0, v1[0], v1[1], color='r', label='v1', angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, v2[0], v2[1], color='b', label='v2', angles='xy', scale_units='xy', scale=1)

v_sum = sum(v1,v2)
plt.quiver(0, 0, v_sum[0], v_sum[1], color='g', label='v1 + v2', angles='xy', scale_units='xy', scale=1)


plt.xlim(0, max(v1[0], v2[0], v_sum[0]) + 1)
plt.ylim(0, max(v1[1], v2[1], v_sum[1]) + 1)
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Двата вектора и векторът от сумата им")

```

* разлика на два вектора в 2Д


```python
plt.quiver(0, 0, v1[0], v1[1], color='r', label='v1', angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, v2[0], v2[1], color='b', label='v2', angles='xy', scale_units='xy', scale=1)

v_diff= subs(v1,v2)
plt.quiver(0, 0, v_diff[0], v_diff[1], color='g', label='v1 - v2', angles='xy', scale_units='xy', scale=1)

plt.xlim(-(max(v1[0], v2[0], v_sum[0])), max(v1[0], v2[0], v_sum[0]) + 1)
plt.ylim(-(max(v1[0], v2[0], v_sum[0])), max(v1[1], v2[1], v_sum[1]) + 1)
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Двата вектора и векторът от разликата им")
```
